```python
# ==============================================================================
# PROJETO AGRINHO - AGRO FORTE, FUTURO SUSTENTÁVEL
# Tema: Equilíbrio entre produção e meio ambiente
# ==============================================================================

import matplotlib.pyplot as plt
import seaborn as sns

# ------------------------------------------------------------------------------
# PARTE 1: Apresentação e Mensagem Inicial
# ------------------------------------------------------------------------------
print("==========================================================")
print("       BEM-VINDO AO SIMULADOR AGRO SUSTENTÁVEL           ")
print("==========================================================")
print("Olá! Este é o meu projeto do Agrinho.")
print("Analisando o equilíbrio entre produção e meio ambiente...\n")

# ------------------------------------------------------------------------------
# PARTE 2: Lógica de Decisão Sustentável (Variáveis e Condições)
# ------------------------------------------------------------------------------
limite_sustentavel = 4000
consumo_fazenda_a = 5000

print("--- AVALIAÇÃO DA FAZENDA TRADICIONAL ---")
if consumo_fazenda_a <= limite_sustentavel:
    print("✅ Parabéns! Consumo de água está equilibrado e sustentável.")
else:
    print("⚠️ Alerta: O consumo passou do limite ambiental! Tente economizar água.")
print("-" * 58)

# ------------------------------------------------------------------------------
# PARTE 3: Geração do Gráfico Moderno Comparativo
# ------------------------------------------------------------------------------
# Configurando o estilo do gráfico moderno
sns.set_theme(style="whitegrid")

# Dados para o comparativo
fazendas = ['Fazenda A\n(Tradicional)', 'Fazenda B\n(Sustentável)', 'Fazenda C\n(Tecnológica)']
consumo_agua = [5000, 2500, 1800]

# Criando a figura do gráfico
plt.figure(figsize=(9, 5.5))

# Desenhando as barras com cores modernas e elegantes
grafico = sns.barplot(
    x=fazendas,
    y=consumo_agua,
    palette="YlGnBu_r",  # Degradê profissional (azul para o verde)
    hue=fazendas,
    legend=False
)

# Adicionando os valores exatos (etiquetas) no topo de cada barra
grafico.bar_label(grafico.containers[0], fmt='%d L/ha', padding=6, fontsize=11, weight='bold', color='#2c3e50')

# Títulos e formatações limpas para o relatório
plt.title('Eficiência Hídrica na Agricultura Sustentável', fontsize=15, fontweight='bold', color='#2c3e50', pad=20)
plt.xlabel('Modelos de Propriedade', fontsize=11, fontweight='bold', color='#34495e', labelpad=12)
plt.ylabel('Consumo de Água (Litros por Hectare)', fontsize=11, fontweight='bold', color='#34495e', labelpad=12)

# Ajustes finais de layout e limites das linhas
plt.ylim(0, 6000)
sns.despine(left=True, bottom=True)  # Remove as bordas pretas da caixa externa

# Exibindo o resultado final na tela
plt.tight_layout()
plt.show()
```
